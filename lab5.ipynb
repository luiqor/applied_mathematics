```python
import sympy as sp

M = (1, 1, 3)
x, y, z = sp.symbols('x y z')

u = sp.sqrt(3*x**2 + 4*y**2 + z**2)
v = sp.sqrt(2*x + 5*y + 3*z)

du_dx = sp.diff(u, x)
du_dx_M = du_dx.subs({x: M[0], y: M[1], z: M[2]})

du_dy = sp.diff(u, y)
du_dy_M = du_dy.subs({x: M[0], y: M[1], z: M[2]})

du_dz = sp.diff(u, z)
du_dz_M = du_dz.subs({x: M[0], y: M[1], z: M[2]})

grad_u = (du_dx_M, du_dy_M, du_dz_M)
print("grad u =", grad_u)

dv_dx = sp.diff(v, x)
dv_dx_M = dv_dx.subs({x: M[0], y: M[1], z: M[2]})

dv_dy = sp.diff(v, y)
dv_dy_M = dv_dy.subs({x: M[0], y: M[1], z: M[2]})

dv_dz = sp.diff(v, z)
dv_dz_M = dv_dz.subs({x: M[0], y: M[1], z: M[2]})

grad_v = (dv_dx_M, dv_dy_M, dv_dz_M)
print("grad v =", grad_v)


gradu_x_gradv = sum(ui * vi for ui, vi in zip(grad_u, grad_v))
calc_gradu = sum(ui**2 for ui in grad_u)
calc_gradv = sum(vi**2 for vi in grad_v)

cos_phi = gradu_x_gradv / (sp.sqrt(calc_gradu) * sp.sqrt(calc_gradv))
print("cos phi =")
sp.pprint(cos_phi)
```

```python
import sympy as sp
x, y = sp.symbols('x y')

z = sp.ln(x*y) + 1/x + 2/y

dz_dx = sp.diff(z, x)
dz_dy = sp.diff(z, y)

# dz/dx = 0  dz/dy = 0
M = sp.solve([dz_dx, dz_dy], (x, y))
x_M = M[0][0]
y_M = M[0][1]
print("M", M)

d2z_dx2 = sp.diff(dz_dx, x)
d2z_dy2 = sp.diff(dz_dy, y)
d2z_dxdy= sp.diff(dz_dx, y)

A = d2z_dx2.subs({x: x_M, y: y_M})

B = d2z_dxdy.subs({x: x_M, y: y_M})

C = d2z_dy2.subs({x: x_M, y: y_M})

delta = A * C - B**2
print("дельта", delta)

if delta > 0 and A >0:
    print("функція 𝑓(𝑥; 𝑦) у т. M має екстремум мінімум")
elif delta > 0 and A<0:
    print("функція 𝑓(𝑥; 𝑦) у т. M має екстремум мінімум")
else:
    print("Якщо ∆= 0 екстремум може бути, а може й не бути. Потрібні додаткові дослідження.")

```
