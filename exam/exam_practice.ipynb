```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.symbols('x')
f = (sp.E**((2 * (x + 2)))) / (2 * (x + 2))

# Find the derivative of the function
dydx = sp.diff(f, x)
#Похідна в спрощеній формі
dy_dx_simplified = sp.simplify(dydx)
print("Похідна в спрощеній формі")
sp.pprint(dy_dx_simplified)

# Convert the symbolic expressions to Python functions for numerical evaluation
f_func = sp.lambdify(x, f, 'numpy')
f_prime_func = sp.lambdify(x, dydx, 'numpy')

# Generate x values for the plot
x_values = np.linspace(-20, 20, 400)
y_values = f_func(x_values)
y_prime_values = f_prime_func(x_values)

# Plot the original function and its derivative
plt.figure(figsize=(8, 6))
#plt.plot(x_values, y_values, label='Original Function')
plt.plot(x_values, y_prime_values, label='Derivative')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.optimize import fsolve

x = sp.symbols('x')

def f(x):
    return x**2 - 5

def g(x):
    return 3 * x

# Find intersection points 
# знаходження таких значень x, при яких різниця між f(x) та g(x) дорівнює нулю, тобто, де обидві функції перетинаються. 
# [] - початкові здогадки x, обов'язкові для fsolve
intersection_points_x = fsolve(lambda x: f(x) - g(x), [-2, 5])
intersection_points_y = [f(x) for x in intersection_points_x]
intersection_points_g = [g(x) for x in intersection_points_x]


# Generate x values for plotting
x_range = np.linspace(-8, 8, 100)

# Plot functions
plt.figure()
plt.plot(x_range, f(x_range), label='f = x^2 - 5', color='purple')
plt.plot(x_range, g(x_range), label='g = 3 * x', color='orange')

# Mark intersection points 
plt.plot(intersection_points_x, f(intersection_points_x), 'o', label='Точки перетину ')

plt.title('Graphs of Functions and Intersection Points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print('Корені системи, підставвивши в y1:')
for x, y in zip(intersection_points_x, intersection_points_y):
    print(f'(x = {x}, y = {y})')

print('\nПеревірка (Корені системи, підставивши в y2):')
for x, y in zip(intersection_points_x, intersection_points_g):
    print(f'(x = {x}, y = {y})')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Визначення функцій
def f(x):
    return x**2 - 5

def g(x):
    return 3 * x

# Побудова графіків
x_values = np.linspace(-3, 3, 400)
plt.plot(x_values, f(x_values), label='f(x) = x^2 - 5')
plt.plot(x_values, g(x_values), label='g(x) = 3*x')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графіки функцій f(x) і g(x)')
plt.show()

# Знаходження коренів системи нелінійних рівнянь
initial_guess = [1.0, 2.0]  # Початкові наближення для коренів

roots = fsolve(lambda x: [f(x[0]) - g(x[0]), g(x[1]) - f(x[1])], initial_guess)

print("Корені системи рівнянь: x1 =", roots[0], ", x2 =", roots[1])
```

```python
import numpy as np
from scipy.optimize import fsolve
from scipy.integrate import quad

# Функція, яку будемо розв'язувати
def equations(vars):
    x, y = vars
    eq1 = 1 - np.sin(x) - y
    eq2 = y - 3/2
    return [eq1, eq2]

# Початкова догадка
initial_guess = [0, 0]

# Розв'язок системи
intersection_points = fsolve(equations, initial_guess)

# Виведення результатів
x, y = intersection_points
print(f'Корені системи: (x = {x}, y = {y})')

# Визначення функцій обмеження для інтеграції
def upper_curve(x):
    return 1 - np.sin(x)

def lower_curve(x):
    return 3/2

# Знаходження площі фігури між кривими
area, _ = quad(lambda x: upper_curve(x) - lower_curve(x), x, np.pi)

print(f'Площа фігури: {area}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

phi = np.linspace(0, 2*np.pi, num=1000)

rl = 1 - np.sin(phi)
r2 = np.ones_like(phi) * 3/2

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.plot(phi, rl, label='rl = 1 + cos(phi)', color='g')
ax.plot(phi, r2, label='r2 = 3/2 (rl <= 3/2)', color='magenta')

# Define the function to integrate
s, _ = quad(lambda phi: (1 - np.sin(phi) - np.ones_like(phi) * 3/2)**2, 0, 2 * np.pi)
s = s / 2
s = round(s, 3)

# Fill the area between curves
ax.fill_between(phi, rl, r2, where=(rl >= r2), color='lightblue', alpha=0.7, label=f'Area: {s}')

ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import sympy as sp

y = sp.symbols('y')

x1 = y**2 - 2*y - 1
x2 = 1 - y

intersection_points = sp.solve(sp.Eq(x1, x2), y)
intersection_y = [float(point.evalf()) for point in intersection_points]

y_vals = np.linspace(min(intersection_y), max(intersection_y), 100)

x1_vals = y_vals**2 - 2*y_vals - 1
x2_vals = 1 - y_vals

plt.plot(x1_vals, y_vals, label='x = y**2 - 2y - 1', color="orange")
plt.plot(x2_vals, y_vals, label='y = 1 - x', color='blue')

area_expr = sp.Abs(x2 - x1)

# Create a numerical function for the area expression
area_func = sp.lambdify(y, area_expr)

# Integrate the numerical function
area, _ = integrate.quad(area_func, min(intersection_y), max(intersection_y))

plt.fill_betweenx(y_vals, x1_vals, x2_vals, color='gray', alpha=0.5, label=f'Area under curve: {round(float(area), 2)}')
plt.title('Integral Plot 2023')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```
